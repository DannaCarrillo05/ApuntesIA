```python
# prompt: cambiar los features ["Latitud","Longitud","Precio_pie_cuadrado"] del df, reemplazar coma por punto

import pandas as pd
# ... (previous code)

# Replace commas with periods in specified columns
for col in ["Latitud", "Longitud", "Precio_pie_cuadrado"]:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)

# Convert columns to numeric, coercing errors to NaN
for col in ["Latitud", "Longitud", "Precio_pie_cuadrado"]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#Now you can handle NaN values
df.dropna(subset=["Latitud", "Longitud", "Precio_pie_cuadrado"], inplace = True)
df[["Latitud","Longitud","Precio_pie_cuadrado"]]
```
