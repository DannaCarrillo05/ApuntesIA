```python
# prompt: multiplicar la dos columnas del arreglo "arreglo" y agregarlo para quedar en la forma (5,3)

import numpy as np

# Crear un arreglo de (5, 2)
arreglo = np.array([[10, 2], [5, 5], [2, 10], [1, 20], [20,1]])

# Multiplicar las dos columnas del arreglo y crear un nuevo arreglo
nuevo_arreglo = arreglo[:, 0] * arreglo[:, 1]

# Reshape the new array to (5, 3) by adding a new column of zeros
nuevo_arreglo = np.column_stack((arreglo, nuevo_arreglo))

nuevo_arreglo
```

```python
# prompt: creame un dataframe con el nombre, el cargo, el salario  y la antiguedad en años, poner 20 filas con datos sinteticos

import pandas as pd
import numpy as np

# Crear datos sintéticos
nombres = np.random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy'], size=20)
cargos = np.random.choice(['Ingeniero', 'Analista', 'Gerente', 'Desarrollador', 'Tester'], size=20)
salarios = np.random.randint(30000, 100000, size=20)
antiguedades = np.random.randint(1, 10, size=20)

# Crear el DataFrame
data = {'Nombre': nombres, 'Cargo': cargos, 'Salario': salarios, 'Antiguedad': antiguedades}
df = pd.DataFrame(data)

# Imprimir el DataFrame
df
```

```python
# prompt: creame un dataframe con datos sinteticos de clientes (nombre,longitud, latitud, valor de compras) la longitud y latitud que sean cercanas a bucaramanga

import pandas as pd
import numpy as np

# Latitud y longitud cercanas a Bucaramanga (aproximadas)
latitud_bucaramanga = 7.125
longitud_bucaramanga = -73.119

# Generar datos sintéticos
num_clientes = 100
nombres = np.random.choice(['Ana', 'Juan', 'Maria', 'Pedro', 'Sofia', 'Luis', 'Laura', 'Carlos', 'Isabel', 'Diego'], size=num_clientes)
latitudes = np.random.normal(loc=latitud_bucaramanga, scale=0.1, size=num_clientes)  # Distribución normal alrededor de Bucaramanga
longitudes = np.random.normal(loc=longitud_bucaramanga, scale=0.1, size=num_clientes)
valor_compras = np.random.randint(10000, 200000, size=num_clientes)

# Crear el DataFrame
data = {'Nombre': nombres, 'Longitud': longitudes, 'Latitud': latitudes, 'Valor_Compras': valor_compras}
df_clientes = pd.DataFrame(data)

# Imprimir el DataFrame
df_clientes
```

```python
import plotly.express as px
fig = px.scatter_mapbox([df_clientes, lat="Latitud", lon="Longitud", color="Valor_Compras", size="Valor_Compras"])
fig.show()
```
